### 4.1 번호 뽑기
- 1~45 사이의 서로 다른 정수, 무작위로 n개 뽑기
- 그리고 그 n개의 리스트를 리턴합니다.
- 참가자의 6개짜리 번호를 뽑는 데 사용할 겁니다! 

```python
import random

n = 6

def generate_numbers(n):
    rand_num_list = []

    while n > 0:
        rand_num = random.randint(1, 45)

        if rand_num not in rand_num_list:
            rand_num_list.append(rand_num)
            n -= 1
    return rand_num_list

print(generate_numbers(n))
```
