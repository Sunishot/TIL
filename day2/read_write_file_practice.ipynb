### 3.5 [실습 - 단어장 만들기](https://www.codeit.kr/topics/making-use-of-python/lessons/2827)

- "영어: 한국어" 이런식으로 단어장(vocabulary.txt)에 정리하도록 하는 프로그램을 만들어볼게요!
- 2가지의 입력을 받습니다.
    - 영어 단어를 입력하세요:
    - 한국어 뜻을 입력하세요:
    - 위 둘 중에 q를 넣으면 바로 프로그램 종료

````python
# 버전 1: append 옵션 사용시 : While을 바깥 껍질에
while True:
    en = input("영어 단어를 입력하세요: ")
    if en == "q":
        break

    ko = input("한국어 뜻을 입력하세요: ")
    if ko == "q":
        break

    with open("vocabulary.txt","a", encoding = "utf-8") as f:
        f.write(f"{en}: {ko}\n")
    print("프로그램 종료 완료!")
````

### 3.6 [실습 - 단어 퀴즈](https://www.codeit.kr/topics/making-use-of-python/lessons/2828)

- 위에서 만든 vocabulary.txt를 바탕으로
- 문제를 내는 프로그램을 만들어 봅시다!
- 순서대로 불러오면서, 사용자가 영어단어를 입력하고,
    - 정답이면, `맞았습니다!`
    - 오답이면, `아쉽습니다. 정답은 OOO입니다.`가 되도록 해봅시다!

````python
with open('vocabulary.txt', 'r', encoding = 'utf-8') as f:
    for line in f:
        data = line.strip().split(": ")
        en, ko = data[0], data[1]

        guess = input(f"{ko}: ")

        if guess == en:
            print(" you are right!")
        else:
            print(f"That's wrong it was {en}.")
````

### 3.7 [실습 - 고급 단어장](https://www.codeit.kr/topics/making-use-of-python/lessons/2829)

- 위에서 만든 단어장을 바탕으로 학생들에 문제를 출제하는 프로그램
- 한국어: (정답) 이런식으로 물어봅니다!
- 틀리면 "아쉽습니다. 정답은 {영어}입니다." 로 말합니다.

    - 3.6이랑 동일한데, `random하게 물어보도록`,
    - 물론, `중복해서 질문`하게될 수도 있어요!
    - `hint.` random이랑 dictionary를 이용할 것!ㅡ

````python
with open('vocabulary.txt', 'r', encoding = 'utf-8') as f:
    vocab = dict()     #이렇게 적어주는게 명시적으로 좋음
    for line in f:
        data = line.strip().split(': ')
        en, ko = data[0], data[1]

    vocab[ko] = en
    keys = list(vocab.keys())
    import random
    while True:
        index = random.randint(0, len(vocab)-1)

        en = list(vocab.keys())[index]
        ko = list(vocab.values())[index]

        guess = input(f"{ko}: ")

        if guess == en:
            print(" you are right!")
        else:
            print(f"That's wrong it was {en}.")
````
